```python
import requests
from bs4 import BeautifulSoup
import os
import urllib.request
import geopandas as gpd
import pandas as pd
import re
#from zipfile import ZipFile
import zipfile
```

```python
response = requests.get("https://data.gcis.nat.gov.tw/od/datacategory", verify=False)
#print(r.status_code)#顯示是否成功
```

```python
 #輸出排版後的HTML內容
soup = BeautifulSoup(response.text, "html.parser")
print(soup.prettify()) 
```

```python
#抓特定區域
a=soup.find(id="cate2")
#抓連結
i=[]
course_links = a.find_all('a', {'href': re.compile('/od/detail;*')})
for link in course_links:
    i.append(link['href'])
```

```python
len(i)
```

```python
#完整連結網址
weball=[]
for com in i:
    j='https://data.gcis.nat.gov.tw'+ com
    #print(weball)
    weball.append(j)
```

```python
#onclick=[]
all_data= pd.DataFrame(columns = ["統一編號","公司名稱","負責人","公司地址","實收資本額","公司狀態","產製日期"])#建立空的datafram
for g in weball:
    r = requests.get(g)
    soup2 = BeautifulSoup(r.text, "html.parser")
    #print(soup2.prettify())  #輸出排版後的HTML內容
    
    #抓下載zip檔的連結
    a=soup2.find_all("a", {"class": "csv"})[0]
    onclick=a.get('onclick')
    onclick=onclick[23:72]
    zip_links='https://data.gcis.nat.gov.tw'+ onclick
    print(zip_links)
    #把zip檔放入本機
    oldname=list(urllib.request.urlretrieve(zip_links))
    oldname=oldname[0][35:46]
    #設置舊文件名（就是路徑+文件名）
    oldname='C:\\Users\\GIS519\\AppData\\Local\\Temp\\'+oldname
    #設置新文件名 
    newname=oldname+'.zip' 
    #用os模塊中的rename方法對文件改名 
    os.rename(oldname,newname)
    #讀取zip
    myzip=zipfile.ZipFile(os.path.join(os.getcwd(), newname))
    name=myzip.namelist()
    for i in name:
        f=myzip.open(i)
    df=pd.read_csv(f)
    print(df)
    f.close()
    myzip.close()
    all_data=all_data.append(df,ignore_index=True)
    #break
```

```python
import shutil
import zipfile

with zipfile.ZipFile( 'tmp2tv44tci.zip', 'r') as zf:
    for fn in zf.namelist():
        right_fn = fn.encode('utf-8').decode('gb18030', errors='ignore')  # 将文件名正确编码
        with open(right_fn, 'wb') as output_file:  # 创建并打开新文件
            with zf.open(fn, 'r') as origin_file:  # 打开原文件
                shutil.copyfileobj(origin_file, output_file)  # 将原文件内容复制到新文件
```
