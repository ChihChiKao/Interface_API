```python
import requests
from bs4 import BeautifulSoup
```

```python
# 下載 Yahoo 首頁內容
r = requests.get('https://tw.news.yahoo.com/search?p=%E8%B1%AC%E7%98%9F&fr=uh3_news_vert_gs&fr2=p%3Anews%2Cm%3Asb')
```

```python
# 確認是否下載成功
if r.status_code == requests.codes.ok:
# 以 BeautifulSoup 解析 HTML 程式碼
 soup = BeautifulSoup(r.text, 'html.parser')
print(soup)
```

```python
# 以 CSS 的 class 抓出各類頭條新聞
for news in soup.select("./u"):
    print(news)
```

```python
if(len(news.select('h3')) > 0):print(news.select('h3')[0].text)
```
